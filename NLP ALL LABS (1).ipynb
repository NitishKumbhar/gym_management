# LAB 3: Write a program to program to tokenize Non-English Languages

This Lab contains two exercises:
1. To tokenize the document (article, news, story, assay) of Indian Language and find the frequency count. 
2. To tokenize the document (article, news, story, assay) of Forighn Language and find the frequency count.

```python
# lab 3
from indicnlp import common
from indicnlp.tokenize import indic_tokenize

# Set up the Indic NLP library for Marathi
common.set_resources_path("path/to/indic_nlp_resources")

# Function to tokenize and count words in a Marathi document
def count_words(text):
    # Tokenize the Marathi text
    tokens = indic_tokenize.trivial_tokenize(text)

    # Count words
    word_freq = {}
    for token in tokens:
        if token in word_freq:
            word_freq[token] += 1
        else:
            word_freq[token] = 1

    return word_freq


marathi_text = """
महाराष्ट्र एक भारतीय राज्य आहे. मराठी म्हणजे मराठी भाषा असलेला जनसंख्येचा एक मराठा जातीच्या व्यक्तीस सुचला पहावा.
मराठी साहित्याच्या क्षेत्रातील ग्रंथालय दिलेल्या सानियाच्या अशी एक बृहत भवनाच्या आच्छादितात आपल्याला शब्दकोश दिला जाईल.
"""

# Count words in the Marathi text
word_frequencies = count_words(marathi_text)

# Print word frequencies
for word, frequency in word_frequencies.items():
    print(f"{word}: {frequency}")

    
    
import nltk
from collections import Counter

# Download the NLTK French tokenizer (if not already downloaded)
nltk.download("punkt")

# Function to tokenize and count words in French
def count_words(text):
    # Tokenize the French text
    tokens = nltk.word_tokenize(text, language="french")

    # Count words
    word_freq = Counter(tokens)

    return word_freq

# Sample French text (replace with your document)
french_text = """
Votre texte en français ici.
"""

# Count words in the French text
word_frequencies = count_words(french_text)

# Print word frequencies
for word, frequency in word_frequencies.items():
    print(f"{word}: {frequency}")
```

```python
# lab 4 and 5
pip install wordnet
import nltk
nltk.download('wordnet')
import nltk
nltk.download('omw-1.4')
from nltk.corpus import wordnet

def get_synonyms(word):
    synonyms = []
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.append(lemma.name())
    return list(set(synonyms))  # Remove duplicates

def get_antonyms(word):
    antonyms = []
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())
    return list(set(antonyms))  # Remove duplicates

if __name__ == "__main__":
    word = input("Enter a word: ")
    
    synonyms = get_synonyms(word)
    antonyms = get_antonyms(word)
    
    if synonyms:
        print(f"Synonyms for '{word}': {', '.join(synonyms)}")
    else:
        print(f"No synonyms found for '{word}'")

    if antonyms:
        print(f"Antonyms for '{word}': {', '.join(antonyms)}")
    else:
        print(f"No antonyms found for '{word}'")

        
```

Discussion: (Need to explain as per the given text)

```python
# lab 8
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize the NLTK stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Sample headlines related to technology trends
posts = [
    "Elon Musk Outlines Coming X Premium Subscription Packages", 
    "Including a Higher Priced Ad Free Option",
    "Meta enhances Threads with keyword search",
    "Snapchat ventures into the world of AI with Dreams",
    "YouTube's humming search for Android",
    "Instagram expands creative possibilities",
    "Pinterest's inclusive body-type technology",
    "TikTok introduces text posts",
    "Meta launches AudioCraft",
    "YouTube is taking Shorts to the next level",
    "TikTok launches Subscriber-Only Videos"
]
# Tokenize and process each post
for post in posts:
    words = word_tokenize(post)
    
    # Stemming
    stemmed_words = [stemmer.stem(word) for word in words]

    # Lemmatization
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

    print("Original Text: ", post)
    print("Stemmed Words: ", " ".join(stemmed_words))
    print("Lemmatized Words: ", " ".join(lemmatized_words))
    print("\n")

```
